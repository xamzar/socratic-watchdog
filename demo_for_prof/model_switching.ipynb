# 🤖 Model Switching

Hot-switch the LLM behind Socrates — no restart needed.
Curated list: DeepSeek, GPT-4o, Claude. Or type any custom model.

```python
%load_ext socratic_watchdog
%socratic_debug on
%socratic_audio off
```

## Show available models

```python
%socratic_model
```

👆 Numbered picker with the current model highlighted.

---

## Default: DeepSeek Chat

### Task — Write a function `is_even(n)` that returns True if n is even

```python
%%socratic
def is_even(n):
    return n // 2 == 0
```

👆 DeepSeek asks a Socratic question about the student's code.

---

## Switch to GPT-4o (by number)

```python
%socratic_model 3
```

### Task — Write a function `is_even(n)` that returns True if n is even

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

👆 Now GPT-4o responds — same task, different model.

---

## Switch to Claude (by number)

```python
%socratic_model 5
```

### Task — Write a function `is_even(n)` that returns True if n is even

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

👆 Claude responds — different voice, same Socratic rules.

---

## Switch back to DeepSeek

```python
%socratic_model 1
```
