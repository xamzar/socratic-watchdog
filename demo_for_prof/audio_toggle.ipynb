# 🔊 Audio Toggle

Socratic Watchdog speaks its questions and praise aloud via TTS.
Toggle audio on/off without restarting the kernel.

```python
%load_ext socratic_watchdog
%socratic_debug on
```

## Current audio state

```python
%socratic_audio
```

## Turn audio OFF

Socrates will still show the amber question box, but no voice.

```python
%socratic_audio off
```

### Task — Write a function `greet(name)` that returns `"Hello, " + name`

```python
%%socratic
def greet(name):
    return "Hello, " + name
```

👆 Correct — praise box appears, but NO audio.

---

## Turn audio ON

Now you'll hear Socrates speak.

```python
%socratic_audio on
```

### Task — Write a function `square(n)` that returns n²

```python
%%socratic
def square(n):
    return n * 2
```

👆 Correct — praise box + audio voice.

---

## Wrong answer with audio ON

Hear Socrates ask a guiding question.

### Task — Write a function `double(n)` that returns n × 2

```python
%%socratic
def double(n):
    return n + 2
```

👆 Wrong logic — amber question box + audio voice.

## Summary

| Command | Effect |
|---------|--------|
| `%socratic_audio on` | Voice + subtitle box |
| `%socratic_audio off` | Subtitle box only, no voice |
| `%socratic_audio` | Show current state |
