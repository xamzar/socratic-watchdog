### Task 1 — Create a dictionary

Create a dictionary `student` with keys `name`, `age`, `grade`.

```python
student = {"name": "Alice", "age": 20, "grade": "A"}
```

### Task 2 — Access by key

Print the value for key `"name"`.

```python
print(student["name"])
```

### Task 3 — Add a key

Add key `"major"` with value `"Computer Science"`, then print the dict.

```python
student["major"] = "Computer Science"
print(student)
```

### Task 4 — Update a value

Change `"age"` to `21`, then print it.

```python
student["age"] = 21
print(student["age"])
```

### Task 5 — Safe lookup with .get()

Look up `"gpa"` using `.get()`.  Return `"Not found"` if missing.

```python
print(student.get("gpa", "Not found"))
```

### Task 6 — Loop over .keys()

Print every key, one per line.

```python
for key in student.keys():
    print(key)
```

### Task 7 — Loop over .values()

Print every value, one per line.

```python
for value in student.values():
    print(value)
```

### Task 8 — Loop over .items()

Print each pair as `"key: value"`.

```python
for key, value in student.items():
    print(f"{key}: {value}")
```
