A Socratic tutor that watches your notebook cells and asks *guiding questions* when you go off-track — it never hands you the answer, and stays silent (with confetti) when you're right.

**The idea:** write the exercise as a normal markdown cell, put your code in a `%%socratic` cell right below it, and Socrates reads the task straight from the markdown above. Run the setup cell once, then go.

## 2. The task lives in a markdown cell

The cell just below is a normal markdown cell describing the exercise. The `%%socratic` cell after it doesn't repeat the task — Socrates reads it from here automatically.

### Task
Write a function `fib(n)` that returns the *n*-th Fibonacci number.

```python
%%socratic
def fib(n):
    return fib(n-1) + fib(n-2)   # oops — no base case!
```

## 3. Add hidden tests for an instant pass/fail

Again the task is just markdown. This time we add a `#Test cases` cell **right below** the `%%socratic` cell — if the code passes every assert, Socrates stays silent and throws confetti, no LLM needed.

### Task
Write `is_even(n)` that returns `True` for even numbers, `False` otherwise.

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

```python
#Test cases
assert is_even(0) == True
assert is_even(2) == True
assert is_even(7) == False
assert is_even(-4) == True
```

## 4. More to try

- **Explicit task override:** `%socratic_task Write a function that ...` sets the task by hand (handy if auto-detect can't see your notebook).
- `%socratic_help` — the full command list
- `%socratic_style brief` — terse questions instead of the playful mentor
- `%socratic_generate_tests` — let the LLM write the hidden tests for you
- `%socratic_debug on` — see per-cell timing and which path ran
