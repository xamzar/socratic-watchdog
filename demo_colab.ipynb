## 2. Off-track code → a guiding question

Set a task, then write code with a bug. Socrates asks *one* question — it won't tell you the fix.

```python
%socratic_task Write a function fib(n) that returns the nth Fibonacci number
```

```python
%%socratic
def fib(n):
    return fib(n-1) + fib(n-2)   # oops — no base case!
```

## 3. Author-written tests → instant pass/fail

Put a `#Test cases` cell **right below** a `%%socratic` cell. If the code passes every assert, Socrates stays silent and throws confetti — no LLM needed. If it fails, the failures sharpen the question.

```python
%socratic_task Write is_even(n) that returns True for even numbers
```

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

```python
#Test cases
assert is_even(0) == True
assert is_even(2) == True
assert is_even(7) == False
assert is_even(-4) == True
```
