# 🧠 Chapter 2 — Test Cases & Caching

Socrates resolves tests in 3 steps:

1. **Cache** — checks on-disk cache (keyed by task hash)
2. **Human-written** — scans the cell below for `#Tests` / `#Test cases`
3. **AI-generate** — calls LLM, caches result for next time

Human-written tests are cached too — they survive kernel restarts.

```python
%unload_ext socratic_watchdog
%load_ext socratic_watchdog
%socratic_debug on
```

## 1. Human-written tests (cell below `#Tests`)

The cell BELOW `%%socratic` has the `#Tests` marker.
Socrates reads it, validates your code against it, and caches it.

### Task — Write a function `is_even(n)` that returns True if n is even

```python
%socratic_clear_cache
```

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

```python
#Tests
assert is_even(0) == True
assert is_even(1) == False
assert is_even(42) == True
assert is_even(-7) == False
```

👆 **Passes** — Socrates sees the `#Tests` cell below, runs them, praises you.
The tests are also cached to disk.

## 2. Cache hit — instant reload

Same task, different code.  No `#Tests` cell this time — but cached tests still fire.

### Task — Write a function `is_even(n)` that returns True if n is even

```python
%%socratic
def is_even(n):
    return n % 2 == 1  # wrong — returns True for odd numbers
```

👆 **Fails** — cached tests catch the bug.  No LLM call, no network.
Socrates asks a guiding question based on the failure.

## 3. AI-generated tests (fallback)

No `#Tests` cell and nothing in cache → LLM generates tests.

```python
%socratic_clear_cache
```

### Task — Write a function `double(x)` that returns x * 2

```python
%%socratic
def double(x):
    return x * 2
```

👆 **AI generated** — LLM creates 4-6 assert statements, caches them.
Second run with the same task hits the cache instantly.
