# 🧠 Chapter 2 — Auto-Generated Tests

With a task, `%%socratic` auto-generates tests (cached by task hash).
First run calls LLM, subsequent runs are instant.

```python
%unload_ext socratic_watchdog
%load_ext socratic_watchdog
%socratic_debug on
%socratic_audio off
```

### Task — Write double(x) that returns x * 2

```python
%socratic_clear_cache
```

```python
%%socratic
def double(x):
    return x * 2
```

👆 First run — LLM generates + caches tests.

### Task — Write double(x) that returns x * 2

```python
%%socratic
def double(x):
    return x + 2
```

👆 Second run — instant cache hit, tests catch the bug.

### Task — Write is_even(n) that returns True if n is even

```python
%socratic_clear_cache
```

```python
%%socratic
def is_even(n):
    return n % 2 == 0
```

### Task — Write is_even(n) that returns True if n is even

```python
%%socratic
def is_even(n):
    return n % 2 == 1
```
