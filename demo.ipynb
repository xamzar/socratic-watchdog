# 🧠 Socratic Watchdog — Demo Notebook

A Socratic (question-based) TTS coding assistant that lives in your notebook.

---

### Two modes

| Mode | How to activate | Best for |
|---|---|---|
| **Explicit task** | `%socratic_task Build a BMI calculator` | Students exploring |
| **Track mode** | `%socratic_task auto` | Pre-made lab tracks — reads task from markdown above |

### New: track author features
- 📝 **Auto-detect** — task is read from the markdown cell above (no retyping)
- 🧪 **`%socratic_tests`** — embed expected test cases for stricter analysis

> *"I cannot teach anybody anything, I can only make them think."* — Socrates

## Part A: Explicit task mode (student sets the goal)

```python
%load_ext socratic_watchdog
```

```python
%socratic_task Write a function that calculates BMI as weight_kg / height_m^2, then classifies it into underweight, normal, overweight, or obese.
```

✅ Correct code — Socrates should stay silent:

```python
%%socratic
def bmi(weight, height):
    return weight / height ** 2

def classify(bmi_value):
    if bmi_value < 18.5:
        return "underweight"
    elif 18.5 <= bmi_value <= 24.9:
        return "normal"
    elif 25 <= bmi_value <= 29.9:
        return "overweight"
    else:
        return "obese"
```

🐛 Buggy code — Socrates should ask a question:

```python
%%socratic_help
def bmi(weight, height):
    return weight / height  # forgot to square!
```

## Part B: Track mode — auto-detect task from markdown

This is how a pre-made lab track would work. The student doesn't retype the task.

```python
%socratic_task auto
```

### Task 1: Palindrome Checker

Write a function `is_palindrome(s)` that returns `True` if the string `s` reads the same forwards and backwards.

The function should:
- Ignore case ("Racecar" → True)
- Ignore spaces ("a man a plan a canal panama" → True)

```python
%%socratic
def is_palindrome(s):
    return s == s[::-1]
```

👆 Socrates should ask a question about case or spaces — because the markdown task explicitly mentions ignoring them, but the student's code doesn't.

### Task 2: Fibonacci

Write a function `fib(n)` that returns the nth Fibonacci number (0-indexed: fib(0)=0, fib(1)=1, fib(2)=1, fib(3)=2, ...).

```python
%%socratic
def fib(n):
    return fib(n-1) + fib(n-2)  # missing base case
```

👆 Socrates should ask about the base case (what stops the recursion).

## Part C: Track mode + test cases (strictest analysis)

### Task 3: Factorial (with pre-assigned tests)

Write a function `factorial(n)` that returns `n!` (the product of all positive integers ≤ n).

factorial(0) should return 1 (by definition).

```python
%socratic_tests
assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(5) == 120
assert factorial(7) == 5040
```

```python
%%socratic
def factorial(n):
    result = 1
    for i in range(1, n):
        result *= i
    return result
```

👆 Bug: `range(1, n)` misses `n` itself. Socrates should:
1. Note the test failure (`factorial(5)` would return 24, not 120)
2. Ask a Socratic question about the loop range

---

## Command reference

| Magic | What it does |
|---|---|
| `%%socratic` | Run a cell with Socratic analysis |
| `%socratic_task ...` | Set explicit task goal |
| `%socratic_task auto` | Auto-read task from markdown above |
| `%socratic_task clear` | Remove the task |
| `%socratic_tests ...` | Set pre-assigned test cases |
| `%socratic_watch on/off` | Watch every cell automatically |
| `%socratic_off` | Stop watching |
| `%socratic_reset` | Clear everything |
| `%socratic_help` | Show full help |
